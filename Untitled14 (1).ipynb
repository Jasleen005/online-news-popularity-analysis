```python
## 🤖 Machine Learning Model
# A simple classification model was created to predict whether an article is popular or not. The target variable was converted into a binary variable where articles with shares greater than 1400 were considered popular.
# A Decision Tree Classifier was used for prediction, achieving an accuracy of around 52%. This shows that predicting news popularity is challenging and depends on multiple complex factors.

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

X = df[['num_imgs', 'num_videos', 'kw_avg_avg']]
y = df['popular']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = DecisionTreeClassifier()
model.fit(X_train, y_train)

model.score(X_test, y_test)
```
